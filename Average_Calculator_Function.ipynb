```python
#Creating the first function titled "Introduction" where it's main purpose is to describe the program in its entirety.
#This function will not be receiving any parameters, user inputs or be returning any values whatsoever.
#It will also explain to the user how to end the program by typing "0 0 0" to exit the program.

def introduction(): #This function will print a message to the user describing the program.
    print("Hello.")
    print()
    print("This program takes three numbers and returns their average. It then compares each number to the average.")
    print()
    print("This program will then display whether each number is greater than, equal to, or less than the average.")
    print() 
    print("It will also count how many numbers are equal to the average.")
    print()
    print("To end this program, enter this combination: 0 0 0")
    print()
    print("Coded by: Adonis Vasquez")
    print()

#The second function will take the three numbers the user types in and finds the average.

def findaverage(num1, num2, num3): #This function will take three parameters. They are "num1", "num2" and "num3" respectively.
    
    #Calculating the average and returning the result. 
    average = (num1 + num2 + num3) / 3
    return average

#The third function will take the first three numbers and compare these numbers to the average that was calculated above.
#Do note that this function will not be returning any values but will be comparing the numbers that the user types in to the resulting average.
#This function will also be counting how many numbers are equal to the average. 

def comparetoavg(num1, num2, num3, avg): #This function takes four parameters. The first three numbers that the user types in and the resulting average.


    equal_count = 0 #Initializing a counter and setting it to 0.

    if num1 == avg: #Comparing the first number to the average and checking to see if its equal to the average.
        print(f"{num1} is equal to the average")
        equal_count += 1
    elif num1 < avg: #Checking to see whether the first number is greater than or less than the average.
        print(f"{num1} is below the average")
    else:
        print(f"{num1} is above the average")
        
    if num2 == avg: #Comparing the second number to the average and checking to see if its equal to the average.
        print(f"{num2} is equal to the average")
        equal_count += 1
    elif num2 < avg: #Checking to see whether the second number is greater than or less than the average.
        print(f"{num2} is below the average")
    else:
        print(f"{num2} is above the average")
        
    if num3 == avg: #Comparing the third number to the average and checking to see if its equal to the average.
        print(f"{num3} is equal to the average")
        equal_count += 1
    elif num3 < avg: #Checking to see whether the third number is greater than or less than the average.
        print(f"{num3} is below the average")
    else:
        print(f"{num3} is above the average")
        
    if equal_count == 1: #Telling the user out of the three numbers, how many were equal to the average. 
        print("1 value equal to the average")
    else:
        print(f"{equal_count} values equal to the average")
        
#The last function will call the introduction function to begin the program.
#Once it calls the introduction function, it will tell the user to type in three numbers of their choosing.
def main(): #The main function doesn't take any parameters, user inputs or returns any values.
  
    introduction() #Calling the introduction function to begin the program.

    set_count = 0 #Initializing the counter to keep track of the program looping itself.

    while True: #Telling the user to enter the three numbers and reminding them of the special combo.
        
        print("Feel free to type in three numbers that are either positive, negative or zero.")
        print()
        print("Do keep in mind that typing 0 for all three numbers will exit the program. Have fun!")
        print()
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        num3 = int(input("Enter the third number: "))

        #Reminding the user what were their original three numbers that they typed in.
        print(f"Your original three numbers were: {num1}, {num2}, {num3}.") 
        
        #Looking for the special combo to exit the program. If combo isn't present then the program will continue.
        #In this case, we are looking to see if all three numbers are equal to 0 0 0.
        if num1 == 0 and num2 == 0 and num3 == 0:
            break
            
        set_count += 1 #This will keep track of how many times the loop was repeated. 
        
        average = findaverage(num1, num2, num3) #Will call the find average function which will calculate the total average of all three numbers.
        print(f"the average is {average:.3f}") #This will output the average in three decimal places.
        
        comparetoavg(num1, num2, num3, average) #Will call the compare average function and have it tell the user the results.
        
        print()
    
    #Will tell the user how many sets of data they have processed.    
    print(f"You have processed {set_count} sets of data. So sad to see you go. Until next time!")
    
if __name__ == "__main__":
    main()
```
